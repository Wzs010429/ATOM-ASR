```python
data = pd.read_csv('data.csv')
data.columns
```

```python
data.head()
```

```python
import opencc
import re
```

```python
# 初始化opencc进行繁体到简体转换
converter = opencc.OpenCC('t2s.json')

# 定义一个函数，进行繁体转简体并删除标点符号
def convert_and_clean(text):
    # 繁体中文转换为简体中文
    simplified_text = converter.convert(text)
    # 删除所有标点符号，只保留中文字符和英文字符
    cleaned_text = re.sub(r'[^\w\s]', '', simplified_text)
    return cleaned_text

# 需要处理的列名
columns_to_process = [
    'sentence',
    'openai/whisper-small',
    'openai/whisper-large-v3',
    'BELLE-2/Belle-whisper-large-v3-zh'
]

# 对指定的列进行处理
for column in columns_to_process:
    data[column] = data[column].apply(convert_and_clean)

```

```python
data.head()
```

```python
import jiwer
import pandas as pd

def calculate_wer(reference, hypothesis):
    return jiwer.wer(reference, hypothesis)

def calculate_cer(reference, hypothesis):
    return jiwer.cer(reference, hypothesis)


# 需要处理的模型列名
model_columns = [
    'openai/whisper-small',
    'openai/whisper-large-v3',
    'BELLE-2/Belle-whisper-large-v3-zh'
]

# 实际的文段列名
reference_column = 'sentence'

# 存储每个模型的WER和CER结果
results = {}

# 计算每个模型的WER和CER
for model in model_columns:
    # 获取参考文本和模型生成的文本
    references = data[reference_column].tolist()
    hypotheses = data[model].tolist()
    
    wer_list = []
    cer_list = []
    
    # 对每一行计算WER和CER
    for ref, hyp in zip(references, hypotheses):
        wer = calculate_wer(ref, hyp)
        cer = calculate_cer(ref, hyp)
        wer_list.append(wer)
        cer_list.append(cer)
    
    # 计算WER和CER的平均值
    average_wer = sum(wer_list) / len(wer_list)
    average_cer = sum(cer_list) / len(cer_list)
    
    # 存储结果
    results[model] = {
        'Average WER': average_wer,
        'Average CER': average_cer
    }

# 打印结果
for model, metrics in results.items():
    print(f"模型: {model}")
    print(f"  平均 WER: {metrics['Average WER']:.4f}")
    print(f"  平均 CER: {metrics['Average CER']:.4f}")
```
